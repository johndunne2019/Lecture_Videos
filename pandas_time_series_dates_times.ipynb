## pandas dealing with dates and times in csv files

### Time series: importing

**Importing csv file containing dates and times and changing the dates and times to a format that pandas can deal with and a simple plot of the data.**

```python
import pandas as pd
```

```python
df = pd.read_csv("http://cli.met.ie/cli/climate_data/webdata/hly4935.csv", skiprows=23, low_memory=False, nrows=1000)
# read the data file straight from url
# skip rows to skip forst 23 rows
# lowmemory=false has to be used (error message was given)
# nrows asks pandas to only read a certain amunt of rows
```

```python
# Change the date column to a Pythonic datetime.
df['datetime'] = pd.to_datetime(df['date'])
# this command adds a new column called date_time containing the dates and times created from the original date columnn. 
# In this new column the dates and times are in a format pandas can work with
```

```python
df
# only 1000 rows outputted as we have entered nrows=1000
```

```python
import seaborn as sns
```

```python
sns.lineplot(x="datetime", y="temp", data=df)
# seaborn.lineplot 
# the straight line is because there was only one data point in 1996 and next was a long time later
```

```python
sns.lineplot(x="datetime", y="temp", data=df[1:])
# [1:] used to skip the first data entry from above
```

```python
%matplotlib inline

# Plot the temperature over time.

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

sns.set(style="darkgrid")

fig, ax = plt.subplots()

sns.lineplot(x="datetime", y="temp", data=df[1:], ax=ax)

# Makes the x axis display well.
# See: https://stackoverflow.com/a/31262531/340324
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%y'))

plt.show()
```

### Time series: simulation

**Simulating data**

```python
# Create a datetime index for a data frame.

# Adapted from: https://pandas.pydata.org/pandas-docs/stable/timeseries.html

# 72 hours starting with midnight Jan 1st, 2011
rng = pd.date_range('1/1/2011', periods=72, freq='H')
```

```python
rng
```

**The above range will be used and an index for a new data set that we will simulate below**

**The Poisson distribution was picked here as it measures the likelihood of an event occurring in a given time period**

```python
# Simulate some hospital admissions data using numpy, using our datetime index.

import numpy as np
ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=["admissions"])
```

```python
ts
```

### Accessing and grouping

```python
# Calculate the mean number of hospital admissions per hour.
ts.mean()
```

```python
# Look at a specific day's data using the index.
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00']
```

```python
# Calculate that day's mean.
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00'].mean()
```

```python
# Use re-sampling to calculate the mean of all days.
ts.resample('D').mean()
```
